```python

path ="../../archivos/metadata.json.gz"
import json
import gzip
def parse(path):
#Abrir gzip
#Open gzip
    g = gzip.open(path, 'r')
    for l in g:
        yield eval(l)

#Convertir a  json estricto
#Turn to strict json
def parse(path):
    g = gzip.open(path, 'r')
    for l in g:
        yield json.dumps(eval(l))
    f = open("output.strict", 'w')
    for l in parse("archivos\metadata.json.gz"):
        f.write(l + '\n')

#Pandas data frame
import pandas as pd
import gzip
def parse(path):
    g = gzip.open(path, 'rb')
    for l in g:
        yield eval(l)
def getDF(path):
    i = 0
    df = {}
    for d in parse(path):
        df[i] = d
        i += 1
    return pd.DataFrame.from_dict(df, orient='index')

df = getDF('../../archivos/metadata.json.gz')
```

```python
#Convertimos df a diccionario para iterarlo
#Turn df to dict and iter
df_dict = df2secure.to_dict('records')
# Iteramos el diccionario para separar la columna salesRank a dos segun clave y contenido del diccionario
# Los diferentes try, except evitan inconvenientes con las anomalias del df

#Iter the dict to expand salesRank column
for i, x in enumerate(df_dict):
    try:
        try:
            rank = int(x["salesRank"])
            category = "N/A"
        except:
            if x["salesRank"] == "N/A" or x["salesRank"] == {} or x['salesRank'] == '':
                category = "N/A"
                rank = "N/A"
            else:
                data = x["salesRank"].split(": ")
                category = data[0]
                rank = data[1]
        x["category"] = category
        x["salesRank"] = rank
    except Exception as e:
        print(f"Failed at register nº{i}.")
        print(e)

df_final = pd.DataFrame.from_dict(df_dict)
print(df_final)
```
